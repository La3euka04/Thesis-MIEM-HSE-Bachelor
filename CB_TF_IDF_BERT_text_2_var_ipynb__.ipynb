```python
import numpy as np
import pandas as pd
from scipy.sparse import coo_matrix

SEED = 42
np.random.seed(SEED)

MIN_USER_INTERACTIONS = 5
MIN_ITEM_INTERACTIONS = 5
K_TEST = 1
K_VAL  = 1

assert MIN_USER_INTERACTIONS >= (K_TEST + K_VAL + 1), "иначе train_inner может стать пустым"

df0 = df.copy()
df0["date"] = pd.to_datetime(df0["date"], errors="coerce")
df0 = (df0
       .dropna(subset=["user_id", "business_id", "date"])
       .sort_values(["user_id", "date"])
)

user_cnt = df0.groupby("user_id")["business_id"].size()
eligible_users = user_cnt[user_cnt >= MIN_USER_INTERACTIONS].index
df_cf = df0[df0["user_id"].isin(eligible_users)].copy()

print("Users after filter:", df_cf["user_id"].nunique())
print("Total interactions:", len(df_cf))
```

## Data split

```python
# test = last K_TEST per user
test_idx = df_cf.groupby("user_id").tail(K_TEST).index
test0_full  = df_cf.loc[test_idx].copy()
train0_full = df_cf.drop(test_idx).copy()

# val = last K_VAL per user inside train0
val_idx = train0_full.groupby("user_id").tail(K_VAL).index
val0_full        = train0_full.loc[val_idx].copy()
train_inner0_full = train0_full.drop(val_idx).copy()

print("train_inner0:", len(train_inner0_full), "val0:", len(val0_full), "test0:", len(test0_full))
print("min interactions per user in train_inner0:",
      int(train_inner0_full.groupby("user_id")["business_id"].size().min()))


train0       = train0_full[["user_id","business_id","date"]].copy()
test0        = test0_full[["user_id","business_id","date"]].copy()
val0         = val0_full[["user_id","business_id","date"]].copy()
train_inner0 = train_inner0_full[["user_id","business_id","date"]].copy()
```

```python
def build_split_objects(train_df, eval_df, min_item_interactions=5):
    item_cnt = train_df.groupby("business_id")["user_id"].size()
    eligible_items = item_cnt[item_cnt >= min_item_interactions].index

    train_f = train_df[train_df["business_id"].isin(eligible_items)].copy()
    eval_f  = eval_df[eval_df["business_id"].isin(eligible_items)].copy()

    users = train_f["user_id"].unique()
    items = train_f["business_id"].unique()
    user2idx = {u:i for i,u in enumerate(users)}
    item2idx = {b:i for i,b in enumerate(items)}

    train_f["u"] = train_f["user_id"].map(user2idx).astype("int32")
    train_f["i"] = train_f["business_id"].map(item2idx).astype("int32")

    eval_f["u"] = eval_f["user_id"].map(user2idx)
    eval_f["i"] = eval_f["business_id"].map(item2idx)

    before = len(eval_f)
    eval_f = eval_f.dropna(subset=["u","i"]).copy()
    eval_f["u"] = eval_f["u"].astype("int32")
    eval_f["i"] = eval_f["i"].astype("int32")
    dropped = before - len(eval_f)

    R = coo_matrix(
        (np.ones(len(train_f), dtype=np.float32),
         (train_f["u"].values, train_f["i"].values)),
        shape=(len(users), len(items))
    ).tocsr()

    user_train_items = train_f.groupby("u")["i"].apply(lambda x: set(x.tolist())).to_dict()
    all_items = np.arange(len(items), dtype=np.int32)

    return {
        "n_users": len(users),
        "n_items": len(items),
        "R": R,
        "train_f": train_f,
        "eval_f": eval_f,
        "eval_u": eval_f["u"].values,
        "eval_i": eval_f["i"].values,
        "user_train_items": user_train_items,
        "all_items": all_items,
        "dropped_cold_start": dropped,
        "eligible_items": eligible_items
    }

val_pack  = build_split_objects(train_inner0, val0,  min_item_interactions=MIN_ITEM_INTERACTIONS)
test_pack = build_split_objects(train0,       test0, min_item_interactions=MIN_ITEM_INTERACTIONS)

print("VAL: n_users/n_items:", val_pack["n_users"], val_pack["n_items"],
      "| eval:", len(val_pack["eval_u"]), "| dropped cold:", val_pack["dropped_cold_start"])
print("TEST: n_users/n_items:", test_pack["n_users"], test_pack["n_items"],
      "| eval:", len(test_pack["eval_u"]), "| dropped cold:", test_pack["dropped_cold_start"])
```

```python
def agg_item_text(train_full: pd.DataFrame, eligible_items: pd.Index) -> pd.Series:
    x = train_full[train_full["business_id"].isin(eligible_items)][["business_id","text"]].copy()
    x["text"] = x["text"].fillna("").astype(str)
    return x.groupby("business_id")["text"].apply(lambda s: " ".join(t for t in s.values if t))



item_text_val  = agg_item_text(train_inner0_full, val_pack["eligible_items"])
item_text_test = agg_item_text(train0_full,       test_pack["eligible_items"])

print("item_text_val items:",  item_text_val.shape[0])
print("item_text_test items:", item_text_test.shape[0])
```

## Metrics

```python
import numpy as np

def _ndcg_from_rank(rank: int) -> float:
    # rank: 0-based
    return 1.0 / np.log2(rank + 2)

def _ap_from_rank(rank: int) -> float:

    return 1.0 / (rank + 1)

def _extract_items(recs):
    """
    Нормализуем вывод recommend_fn:
    - если вернул np.array/list -> это items
    - если вернул (items, scores) -> берём items
    """
    if recs is None:
        return np.array([], dtype=np.int32)
    if isinstance(recs, tuple) and len(recs) >= 1:
        recs = recs[0]
    return np.asarray(recs, dtype=np.int32)

def evaluate_next_item(eval_u, eval_i, recommend_fn, ks=(5,10,20,50), name="model"):
    """
    Оценка next-item (по одному target на пользователя).
    ВАЖНО: ничего дополнительно из eval не выкидываем (в т.ч. если target item был seen ранее).
    """
    ks = tuple(sorted(ks))
    maxK = ks[-1]
    n = len(eval_u)
    if n == 0:
        print(f"[{name}] empty eval set")
        return {k: (0.0, 0.0, 0.0) for k in ks}

    rec = {k: 0.0 for k in ks}
    ndcg = {k: 0.0 for k in ks}
    mp = {k: 0.0 for k in ks}
    short_lists = 0

    for u, pos in zip(eval_u, eval_i):
        items = _extract_items(recommend_fn(int(u), int(maxK)))

        if len(items) < maxK:
            short_lists += 1


        rpos = None
        for r, it in enumerate(items):
            if int(it) == int(pos):
                rpos = r
                break

        for k in ks:
            if rpos is None or rpos >= k:
                continue
            rec[k] += 1.0
            ndcg[k] += _ndcg_from_rank(rpos)
            mp[k] += _ap_from_rank(rpos)

    out = {k: (rec[k]/n, ndcg[k]/n, mp[k]/n) for k in ks}
    print(f"[{name}] short rec lists: {short_lists}/{n} ({short_lists/n:.3f})")
    return out

def print_metrics(name, out, ks=(5,10,20,50)):
    print(f"\n=== {name} ===")
    for k in ks:
        r, n, m = out[k]
        print(f"K={k:2d} | Recall@{k}: {r:.4f} | NDCG@{k}: {n:.4f} | MAP@{k}: {m:.4f}")
```

## Feature engeneering

```python

need_item_cols = [c for c in ["business_id","categories","attributes","stars_business","review_count","is_open"]
                  if c in df_cf.columns]

items_meta = (df_cf[need_item_cols]
              .sort_values(["business_id"])
              .groupby("business_id", as_index=False)
              .first())
```

```python
import re, ast

_re_tok = re.compile(r"[^a-z0-9]+", re.IGNORECASE)

def _clean_token(s: str) -> str:
    s = str(s).strip().lower()
    s = s.replace("&", " and ")
    s = _re_tok.sub("_", s).strip("_")
    return s

def _cat_tokens(cat):
    if pd.isna(cat) or cat is None:
        return []
    parts = [p.strip() for p in str(cat).split(",") if p.strip()]
    return [f"cat_{_clean_token(p)}" for p in parts]

def _safe_literal_eval(x):
    if isinstance(x, dict):
        return x
    if pd.isna(x) or x is None:
        return {}
    s = str(x)
    # убираем префиксы u'...' / u"..."
    s = s.replace("u'", "'").replace('u"', '"')
    try:
        return ast.literal_eval(s)
    except Exception:
        return {}

def _norm_scalar(v):
    if v is None or (isinstance(v, float) and np.isnan(v)):
        return None
    if isinstance(v, bool):
        return "true" if v else "false"
    s = str(v).strip()
    s = s.replace("u'", "'").replace('u"', '"')

    if len(s) >= 2 and ((s[0] == s[-1] == "'") or (s[0] == s[-1] == '"')):
        s = s[1:-1].strip()
    return s if s else None

def _attr_tokens(attr):
    d = _safe_literal_eval(attr)
    out = []
    for k, v in (d or {}).items():
        if v is None:
            continue
        key = _clean_token(k)


        if isinstance(v, str) and v.strip().startswith("{") and v.strip().endswith("}"):
            v2 = _safe_literal_eval(v)
            if isinstance(v2, dict):
                v = v2

        if isinstance(v, dict):
            for kk, vv in v.items():
                vv = _norm_scalar(vv)
                if vv is None:
                    continue
                if str(vv).lower() in ("true", "false"):
                    if str(vv).lower() == "true":
                        out.append(f"attr_{key}_{_clean_token(kk)}")
                else:
                    out.append(f"attr_{key}_{_clean_token(kk)}_{_clean_token(vv)}")
        else:
            vv = _norm_scalar(v)
            if vv is None:
                continue
            if str(vv).lower() in ("true", "false"):
                if str(vv).lower() == "true":
                    out.append(f"attr_{key}")
            else:
                out.append(f"attr_{key}_{_clean_token(vv)}")
    return out

def build_item_docs_for_pack(pack, items_meta, item_text_series):
    """
    Возвращает:
      idx2biz: np.array business_id в порядке i=0..n_items-1 (строго как в pack)
      docs:    list[str] документов (categories+attributes+review_text_agg + числовые токены)
    """
    idx2biz = (pack["train_f"][["i","business_id"]]
               .drop_duplicates("i")
               .sort_values("i")["business_id"].to_numpy())

    meta = items_meta.set_index("business_id")
    docs = []

    for b in idx2biz:
        row = meta.loc[b] if b in meta.index else None

        cats = _cat_tokens(row["categories"]) if row is not None and "categories" in meta.columns else []
        attrs = _attr_tokens(row["attributes"]) if row is not None and "attributes" in meta.columns else []

        extra = []
        if row is not None:
            if "stars_business" in meta.columns and pd.notna(row["stars_business"]):

                sb = int(round(float(row["stars_business"]) * 10))
                extra.append(f"num_starsbiz_{sb}")
            if "review_count" in meta.columns and pd.notna(row["review_count"]):

                rc = int(row["review_count"])
                bin_rc = int(np.clip(np.floor(np.log10(rc + 1) * 10), 0, 50))
                extra.append(f"num_reviewcountbin_{bin_rc}")
            if "is_open" in meta.columns and pd.notna(row["is_open"]):
                extra.append("flag_open" if int(row["is_open"]) == 1 else "flag_closed")

        rev = item_text_series.get(b, "")

        docs.append(" ".join(cats + attrs + extra) + " " + str(rev))

    return idx2biz, docs


idx2biz_val,  item_docs_val  = build_item_docs_for_pack(val_pack,  items_meta, item_text_val)
idx2biz_test, item_docs_test = build_item_docs_for_pack(test_pack, items_meta, item_text_test)

print("VAL docs:", len(item_docs_val), "TEST docs:", len(item_docs_test))
```

```python
import numpy as np
import pandas as pd

def _bucket_review_count(x):
    try:
        x = int(x)
    except:
        return None
    if x <= 10: return "review_count:0-10"
    if x <= 50: return "review_count:11-50"
    if x <= 200: return "review_count:51-200"
    if x <= 1000: return "review_count:201-1000"
    if x <= 5000: return "review_count:1001-5000"
    return "review_count:5000+"

def _nl_categories(cat):
    if pd.isna(cat) or cat is None:
        return ""
    parts = [p.strip() for p in str(cat).split(",") if p.strip()]
    return "Categories: " + "; ".join(parts) if parts else ""

def _nl_attributes(attr):
    d = _safe_literal_eval(attr)  # как у тебя выше
    if not d:
        return ""
    chunks = []
    for k, v in d.items():
        if v is None:
            continue

        if isinstance(v, str) and v.strip().startswith("{") and v.strip().endswith("}"):
            v2 = _safe_literal_eval(v)
            if isinstance(v2, dict):
                for kk, vv in v2.items():
                    vv = _norm_scalar(vv)  # как у тебя выше
                    if vv is None:
                        continue
                    if str(vv).lower() == "true":
                        chunks.append(f"{k}: {kk}")
                    elif str(vv).lower() == "false":
                        continue
                    else:
                        chunks.append(f"{k}: {kk}={vv}")
                continue

        vv = _norm_scalar(v)
        if vv is None:
            continue
        if str(vv).lower() == "true":
            chunks.append(f"{k}")
        elif str(vv).lower() == "false":
            continue
        else:
            chunks.append(f"{k}: {vv}")

    return "Attributes: " + "; ".join(chunks) if chunks else ""

def build_item_docs_nl_for_pack(pack, items_meta, item_text_series=None):
    """
    item_text_series: dict-like {business_id -> aggregated_review_text}
      (например item_text_val / item_text_test из твоего leakage-safe пайплайна)
    """
    idx2biz = (pack["train_f"][["i","business_id"]]
               .drop_duplicates("i").sort_values("i")["business_id"].to_numpy())

    meta = items_meta.set_index("business_id")
    docs = []

    for b in idx2biz:
        parts = []
        if b in meta.index:
            row = meta.loc[b]

            if "name" in meta.columns and pd.notna(row.get("name", None)):
                parts.append(f"Name: {row['name']}")
            parts.append(_nl_categories(row.get("categories", None)))
            parts.append(_nl_attributes(row.get("attributes", None)))

            if "stars_business" in meta.columns and pd.notna(row.get("stars_business", None)):
                parts.append(f"Rating: {float(row['stars_business']):.1f}")
            if "review_count" in meta.columns:
                rc = _bucket_review_count(row.get("review_count", None))
                if rc: parts.append(rc)
            if "is_open" in meta.columns and pd.notna(row.get("is_open", None)):
                parts.append("Open" if int(row["is_open"]) == 1 else "Closed")

        # добавляем агрегированный текст отзывов (если передан)
        if item_text_series is not None:
            rev = item_text_series.get(b, "")
            if isinstance(rev, float) and np.isnan(rev):
                rev = ""
            if str(rev).strip():
                parts.append("Review: " + str(rev))

        doc = ". ".join([p for p in parts if p and str(p).strip()])
        docs.append(doc)

    return idx2biz, docs

# ВАЖНО: item_text_val / item_text_test должны быть без утечки
idx2biz_val,  item_docs_val  = build_item_docs_nl_for_pack(val_pack,  items_meta, item_text_series=item_text_val)
idx2biz_test, item_docs_test = build_item_docs_nl_for_pack(test_pack, items_meta, item_text_series=item_text_test)

print("VAL docs:", len(item_docs_val), "TEST docs:", len(item_docs_test))
print("Example doc:", item_docs_val[0])
```

```python
import gc
import numpy as np
import torch
from sentence_transformers import SentenceTransformer

KS = (5, 10, 20, 50)

def encode_item_embeddings(item_docs, model_name="sentence-transformers/all-MiniLM-L6-v2",
                           device="cuda", batch_size=256, normalize=True):
    """
    Возвращает item_emb: np.ndarray shape (n_items, d), float32.
    normalize=True -> косинус = dot, можно ранжировать по dot().
    """
    model = SentenceTransformer(model_name, device=device)
    emb = model.encode(
        item_docs,
        batch_size=int(batch_size),
        show_progress_bar=True,
        convert_to_numpy=True,
        normalize_embeddings=bool(normalize),
    ).astype(np.float32)
    return emb


def build_user_profiles(pack, item_emb):
    """
    pack["user_train_items"]: dict u -> set(i)
    Возвращает user_emb: np.ndarray (n_users, d)
    """
    n_users = pack["n_users"]
    d = item_emb.shape[1]
    user_emb = np.zeros((n_users, d), dtype=np.float32)

    for u in range(n_users):
        seen = pack["user_train_items"].get(u, None)
        if not seen:
            continue
        idx = np.fromiter(seen, dtype=np.int32)
        user_emb[u] = item_emb[idx].mean(axis=0)

    # Нормализуем профили, чтобы dot был косинусом (если item_emb нормализован)
    norms = np.linalg.norm(user_emb, axis=1, keepdims=True)
    norms[norms == 0] = 1.0
    user_emb = (user_emb / norms).astype(np.float32)
    return user_emb


def make_bert_cb_recommender(pack, user_emb, item_emb, filter_seen=True, cache_topk=0):
    """
    Исправление утечки ОЗУ:
    - НЕ кешируем полный scores (n_items,) для каждого пользователя.
    - Опционально можно кешировать только top cache_topk (индексы), если cache_topk > 0.

    Для оффлайн-оценки (val/test) рекомендую cache_topk=0.
    """
    user_train_items = pack["user_train_items"]
    cache = {}  # u -> np.int32 топ-индексы (если cache_topk>0)

    def recommend_fn(u: int, K: int):
        K = int(K)
        if K <= 0:
            return np.array([], dtype=np.int32)

        # Если включён кеш топа и он уже посчитан
        if cache_topk and (u in cache):
            top = cache[u]
            return top[: min(K, top.shape[0])]

        # Считаем scores на лету (не сохраняем целиком)
        scores = (user_emb[u] @ item_emb.T).astype(np.float32, copy=False)

        if filter_seen:
            seen = user_train_items.get(u, None)
            if seen:
                # важно: делаем copy, чтобы безопасно проставить -inf
                scores = scores.copy()
                scores[list(seen)] = -np.inf

        # Считаем топ
        KK = K
        if cache_topk:
            KK = min(max(K, int(cache_topk)), scores.shape[0])
        else:
            KK = min(K, scores.shape[0])

        top = np.argpartition(-scores, KK - 1)[:KK]
        top = top[np.argsort(-scores[top])].astype(np.int32)

        if cache_topk:
            cache[u] = top  # сохраняем только индексы, не scores

        return top[: min(K, top.shape[0])]

    return recommend_fn


MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"
DEVICE = "cuda"
BATCH = 128

# ===== VAL =====
item_emb_val = encode_item_embeddings(
    item_docs_val, model_name=MODEL_NAME, device=DEVICE, batch_size=BATCH, normalize=True
)
user_emb_val = build_user_profiles(val_pack, item_emb_val)

# cache_topk=0 -> вообще без кеша (лучше для evaluate)
rec_val = make_bert_cb_recommender(val_pack, user_emb_val, item_emb_val, filter_seen=True, cache_topk=0)

val_out = evaluate_next_item(val_pack["eval_u"], val_pack["eval_i"], rec_val, ks=KS, name="CB-BERT@VAL")
print_metrics("CB-BERT@VAL", val_out, ks=KS)

# Освобождаем память перед TEST
del item_emb_val, user_emb_val, rec_val, val_out
gc.collect()
if DEVICE.startswith("cuda") and torch.cuda.is_available():
    torch.cuda.empty_cache()

# ===== TEST =====
item_emb_test = encode_item_embeddings(
    item_docs_test, model_name=MODEL_NAME, device=DEVICE, batch_size=BATCH, normalize=True
)
user_emb_test = build_user_profiles(test_pack, item_emb_test)

rec_test = make_bert_cb_recommender(test_pack, user_emb_test, item_emb_test, filter_seen=True, cache_topk=0)

test_out = evaluate_next_item(test_pack["eval_u"], test_pack["eval_i"], rec_test, ks=KS, name="CB-BERT@TEST")
print_metrics("CB-BERT@TEST", test_out, ks=KS)
```

```python
import numpy as np
import pandas as pd

def build_business_sentiment_token_map(
    train_df,               # train_inner0 для VAL или train0 для TEST (user_id,business_id,date)
    df_with_text,           # df_cf (обязательно содержит text)
    model_name="distilbert-base-uncased-finetuned-sst-2-english",
    device=0,               # 0 для GPU, -1 для CPU
    batch_size=64,
    max_len=192,
    max_reviews_per_biz=10, # ограничение скорости
    bins=10                 # количество бинов для токена
):
    """
    Возвращает:
      biz2tok:  dict business_id -> "sent_bin_XX"
      biz2mean: dict business_id -> mean_sent in [-1,1]
    """
    # 1) достаем тексты только из train (чтобы без утечки)
    cols = ["user_id","business_id","date"]
    m = (train_df[cols]
         .merge(df_with_text[cols + ["text"]], on=cols, how="left"))
    before = len(m)
    m = m.dropna(subset=["text"])
    m["text"] = m["text"].astype(str)
    m = m[m["text"].str.strip() != ""]
    after = len(m)

    print(f"[sent] merge rows: {before} -> with_text: {after} (missing/empty: {before-after})")
    print(f"[sent] unique businesses in text rows: {m['business_id'].nunique()}")

    # 2) ограничим число отзывов на бизнес (для скорости)
    m = m.sort_values(["business_id","date"])
    m = m.groupby("business_id").head(int(max_reviews_per_biz)).copy()
    print(f"[sent] after cap {max_reviews_per_biz}/biz: rows={len(m)} | biz={m['business_id'].nunique()}")

    # 3) sentiment pipeline
    from transformers import pipeline
    clf = pipeline(
        task="sentiment-analysis",
        model=model_name,
        device=device
    )

    texts = m["text"].tolist()
    bizs  = m["business_id"].tolist()

    sums = {}
    cnts = {}

    def _to_signed_score(res):
        # res может быть dict (label/score) или list[dict] (all scores)
        if isinstance(res, list):
            d = {str(x["label"]).upper(): float(x["score"]) for x in res}
            # POSITIVE/NEGATIVE или LABEL_1/LABEL_0
            if "POSITIVE" in d:
                pos, neg = d["POSITIVE"], d.get("NEGATIVE", 1.0 - d["POSITIVE"])
            elif "LABEL_1" in d:
                pos, neg = d["LABEL_1"], d.get("LABEL_0", 1.0 - d["LABEL_1"])
            else:
                # fallback: берем максимум как pos
                pos = max(d.values())
                neg = 1.0 - pos
        else:
            lab = str(res["label"]).upper()
            sc = float(res["score"])
            # стараемся восстановить pos_prob
            if lab in ("POSITIVE","LABEL_1"):
                pos = sc
            else:
                pos = 1.0 - sc
            neg = 1.0 - pos
        signed = float(pos - neg)  # [-1,1]
        return signed

    # батчевый прогон
    for start in range(0, len(texts), batch_size):
        batch = texts[start:start+batch_size]
        out = clf(batch, truncation=True, max_length=int(max_len))
        # out: list[dict]
        for b, r in zip(bizs[start:start+batch_size], out):
            s = _to_signed_score(r)
            sums[b] = sums.get(b, 0.0) + s
            cnts[b] = cnts.get(b, 0) + 1

        if (start // batch_size) % 50 == 0:
            print(f"[sent] processed {min(start+batch_size, len(texts))}/{len(texts)}")

    biz2mean = {b: (sums[b] / cnts[b]) for b in sums.keys()}

    # 4) биннинг -> токен
    edges = np.linspace(-1.0, 1.0, int(bins) + 1)  # например 10 бинов
    def _bin_token(x):
        # digitize: 1..bins, приведем к 0..bins-1
        k = int(np.digitize([x], edges, right=False)[0] - 1)
        k = max(0, min(k, bins-1))
        return f"sent_bin_{k:02d}"

    biz2tok = {b: _bin_token(v) for b, v in biz2mean.items()}

    # промежуточная статистика
    vals = np.array(list(biz2mean.values()), dtype=np.float32)
    print("[sent] mean score stats:",
          f"n_biz={len(vals)} | mean={vals.mean():.3f} | std={vals.std():.3f} | "
          f"p10={np.quantile(vals,0.1):.3f} | p50={np.quantile(vals,0.5):.3f} | p90={np.quantile(vals,0.9):.3f}")

    tok_counts = pd.Series(list(biz2tok.values())).value_counts().sort_index()
    print("[sent] token distribution:")
    print(tok_counts)

    return biz2tok, biz2mean


# ===== VAL: sentiment только из train_inner0 (без утечки в val0) =====
biz2tok_val, biz2mean_val = build_business_sentiment_token_map(
    train_df=train_inner0,
    df_with_text=df_cf,
    device=0,                # GPU
    batch_size=128,
    max_len=192,
    max_reviews_per_biz=10,  # ускорение
    bins=10
)

# ===== TEST: sentiment только из train0 (без утечки в test0) =====
biz2tok_test, biz2mean_test = build_business_sentiment_token_map(
    train_df=train0,
    df_with_text=df_cf,
    device=0,                # GPU
    batch_size=128,
    max_len=192,
    max_reviews_per_biz=10,
    bins=10
)

# быстрый sanity: посмотрим пару значений
print("VAL sample:", list(biz2tok_val.items())[:5])
print("TEST sample:", list(biz2tok_test.items())[:5])
```

```python
import numpy as np
from scipy.sparse import diags
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize


def train_cb_tfidf(pack, item_docs, vec_params):
    """
    pack: val_pack или test_pack (из build_split_objects)
    item_docs: list[str] длины pack["n_items"] в порядке item index i=0..n_items-1
    vec_params: параметры TfidfVectorizer
    """
    vec = TfidfVectorizer(**vec_params)
    X_items = vec.fit_transform(item_docs).astype(np.float32)  # (n_items, n_terms)


    R = pack["R"].astype(np.float32)  # (n_users, n_items)
    prof = R @ X_items  # (n_users, n_terms)

    cnt = np.asarray(R.getnnz(axis=1)).astype(np.float32)
    cnt[cnt == 0] = 1.0
    prof = diags(1.0 / cnt) @ prof
    prof = normalize(prof, norm="l2", axis=1)

    model = {"vec": vec, "X_items": normalize(X_items, norm="l2", axis=1), "prof": prof}
    return model

def make_recommender(pack, model, filter_seen=True):
    X_items = model["X_items"]   # (n_items, n_terms), L2-норм
    prof = model["prof"]         # (n_users, n_terms), L2-норм

    user_train_items = pack["user_train_items"]
    cache = {}  # scores cache per user

    def recommend_fn(u: int, K: int):
        if u not in cache:
            scores = (prof[u] @ X_items.T).toarray().ravel()  # (n_items,)
            if filter_seen:
                seen = user_train_items.get(u, None)
                if seen:
                    scores[list(seen)] = -np.inf
            cache[u] = scores
        scores = cache[u]

        if K <= 0:
            return np.array([], dtype=np.int32)
        K = min(K, scores.shape[0])


        idx = np.argpartition(-scores, K-1)[:K]
        idx = idx[np.argsort(-scores[idx])]
        return idx.astype(np.int32)

    return recommend_fn
```
